#pyne.xs.ENDFDataSource

This is a class to grab discretized reaction data from an ENDF data source. You can initialize it:

```python
endfds = ENDFDataSource(str_library)
```

Then you can get the discretized data. This returns the group cross-sections of the reaction in order of decreasing energy:

```python
nuc, rx, nuc_i = 40000, 600, 40040
endfds._load_reaction(nuc, rx, nuc_i)
endfds.discretize(nuc, rx, nuc_i)
rxdata = endfds.rxcache[nuc, rx, nuc_i]
grp_struct = rxdata['src_group_struct']
print "group boundaries: {}".format(rxdata['intpoints'])
print "interpolation schemes by energy group: {}".format(dict(zip(range(4, -1, -1), rxdata['intschemes'])))
#print "interpolation energies: {}".format(rxdata['Eint'])
#print "interpolation xs's in group E_0: {}".format(rxdata['xs'])
print "group cross-sections: {}".format(rxdata['dst_sigma'])
```
